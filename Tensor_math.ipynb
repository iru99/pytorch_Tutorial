**Tensor Math Operation**

1) Add, Substract, Multiply, Devide, Reminders, Exponents
2) Shorthand and uh, longhand
3) Reassignment

```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
tensor_a + tensor_b
```

```python
# Addition longhand using .add

torch.add(tensor_a, tensor_b)
```

```python
# substract

tensor_b - tensor_a
```

```python
# .sub function

torch.sub(tensor_b, tensor_a)
```

```python
tensor_a * tensor_b
```

```python
# .mul function

torch.mul(tensor_a, tensor_b)
```

```python
tensor_b / tensor_a
```

```python
# .dev function

torch.div(tensor_b, tensor_a)
```

```python
tensor_b % tensor_a
```

```python
# .reminder function

torch.remainder(tensor_b, tensor_a)
```

```python
# Expornents / Power

torch.pow(tensor_a, tensor_b)
```

```python
# Another way to write

tensor_a.add(tensor_b)
```

```python
# Reassignment

tensor_a = tensor_a + tensor_b
tensor_a
```

```python
tensor_a.add_(tensor_b)
```
