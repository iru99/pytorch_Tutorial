Tensors



*   A torch. Tensor is a multi-dimentional matrix containing elements of a single data type
*   Similar to nympy arrays, but full of fun things that make them work better on GPU's
*   default data type of float32
*   more suitable for deep learning than numpy array







```python
# List

my_list = [[1,2,3],[4,5,6]]
my_list
```

```python
# Numpy Arrays

import numpy as np

np1 = np.random.rand(3,4)
np1
```

```python
# Numpy data type

np1.dtype
```

```python
# Tensor
import torch

tensor_2d = torch.randn(3,4)
tensor_2d
```

```python
# Tensor Operations

my_torch = torch.arange(10)
my_torch
```

```python
# Reshape and View

my_torch = my_torch.reshape(2,5)
my_torch
```

```python
# reshape if we dont know the number of items using -1

my_torch2 = torch.arange(10)
my_torch2
```

```python
my_torch2 = my_torch2.reshape(2,-1)
my_torch2
```

```python
my_torch3 = torch.arange(10)
my_torch3
```

```python
my_torch4 = my_torch.view(2,5)
my_torch4
```

```python
# with reshape and view, they will update

my_torch5 = torch.arange(10)
my_torch5
```

```python
my_torch6 = my_torch5.reshape(2,5)
my_torch6
```

```python
my_torch5[1] = 20
my_torch5
```

```python
my_torch6

# when you made changes they will automatically update in reshape and view
```

**Slices**

```python
my_torch7 = torch.arange(10)
my_torch7
```

```python
# grab a specific item
my_torch7[6]
```

```python
# Grab a slice

my_torch8 = my_torch7.reshape(5,2)
my_torch8
```

```python
my_torch8[:,1]
```

```python
# return a column
my_torch8[:,1:]
```
